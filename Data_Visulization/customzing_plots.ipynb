```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# loading the dimond dataset
df = sns.load_dataset('diamonds')
```

```python
# Taking the first 5000 rows of the dataset
# and displaying the first few rows of the dataset
print(f"Number of rows before sampling: {df.shape[0]}")
print(f"Number of columns before sampling: {df.shape[1]}")
print(f"{50*'-'}")
df = df.sample(5000)
print(f"Number of rows after sampling: {df.shape[0]}")
print(f"Number of columns after sampling: {df.shape[1]}")

# Display the first few rows of the dataset
df.head()
```

# Setting Title

```python
# setting tile
sns.scatterplot(x='carat', y='price', data=df, hue='cut')
plt.title('Scatter plot of carat vs price')
plt.show()
```

```python
# seting the x labal
sns.scatterplot(x='carat', y='price', data=df)
plt.xlabel('Carat of Diamond')  # Label for x-axis
plt.ylabel('Price ($)')  # Label for y-axis
plt.show()
```

### Customizing the Legend

```python

sns.scatterplot(x='carat', y='price', hue='cut', data=df)
```

### 4. Customizing Ticks


```python
sns.histplot(df['price'], bins=10, edgecolor='blue')
plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)  # Change the font size of y-axis labels
plt.show()
```

### 5. Adding a Grid

```python
# sns.set_style('darkgrid')  # Use a grid background
sns.scatterplot(x='carat', y='price', data=df)
plt.grid(True, linestyle='--', alpha=0.5)  # Customize grid lines
plt.show()
```

### 6. Customizing Spines

```python
# Customizing Spines
sns.boxplot(x='carat', y='price', data=df)
plt.gca().spines['top'].set_visible(False)  # Hide the top spine
plt.gca().spines['right'].set_visible(False)  # Hide the right spine
plt.gca().spines['left'].set_color('red')  # Change the color of the left spine
plt.gca().spines['bottom'].set_color('blue')  # Change the color of the bottom spine
plt.show()
```

### 7. Figure size

```python
# Figure size
plt.figure(figsize=(10, 6))  # Set the figure size
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Scatter plot of carat vs price', fontsize=16)  # Set title with font size
plt.xlabel('Carat of Diamond', fontsize=14)  # Set x-axis label with font size
plt.ylabel('Price ($)', fontsize=14)  # Set y-axis label with font size
plt.show()
```

### 8. Creating Subplots

```python
# Creating Subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Create a 2x2 grid of subplots
sns.scatterplot(x='carat', y='price', data=df, ax=axes[0, 0])  # Top-left subplot
sns.histplot(df['price'], bins=10, edgecolor='blue', ax=axes[0, 1])  # Top-right subplot
sns.boxplot(x='cut', y='price', data=df, ax=axes[1, 0])  # Bottom-left subplot

sns.scatterplot(x='carat', y='price', hue='cut', data=df, ax=axes[1, 1])  # Bottom-right subplot

plt.tight_layout()  # Adjust layout to prevent overlap

plt.show()  # Show the plots
```

### 10. Customizing Color Palettes

```python
#Customizing Color Palettes
sns.set_palette('pastel')  # Set a pastel color palette
sns.scatterplot(x='carat', y='price', data=df, hue='cut')
plt.title('Scatter plot of carat vs price with pastel palette')
plt.show()
```

```python
figure = plt.figure(figsize=(15, 4))
sns.set_palette('muted') # muted, deep, pastel, bright, dark, colorblind
sns.boxplot(x='cut', y='price', data=df, hue='clarity')
plt.title('Box Plot with Muted Color Palette')
plt.show()
```
